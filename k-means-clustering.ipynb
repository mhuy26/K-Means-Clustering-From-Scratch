```python
import pandas as pd
import numpy as np
```

```python
players = pd.read_csv('players_22.csv')
players
```

```python
features = ['overall', 'potential', 'value_eur', 'wage_eur', 'age']
```

```python
players = players.dropna(subset=features)
players
```

```python
data = players[features].copy()
data 
```

<h1> STEPS OF K-MEAN CLUSTERING
<h3>1. Scale data
<h3>2. Initialize random centroids
<h3>3. Lable each data point
<h3>4. Update centroids
<h3>5. Repeat 3+4 until centroids stop changing


<h2>Scale data - Min-Max Scaling (Normalization):

```python
data = ((data - data.min())/ (data.max() - data.min()) * 9 + 1)
data.describe()
```

<h2> Initialize random centroids

```python
def rand_centroids(data, k):
    centroids = []

    for i in range(k):
        centroid = data.apply(lambda x : float(x.sample()))
        centroids.append(centroid)

    return pd.concat(centroids, axis = 1) 

centroids = rand_centroids(data, 5)
centroids 
 
```

<h1> Lable each data points (players) according to the cluster centers
<h4> The function looks at each player and compute the distance between that player and the cluster center
<h4> and return the closest cluster center to the data point (player)

```python
#calculate the distance:
def get_lable(data, centroids):
    distances = centroids.apply(lambda x: np.sqrt(((data - x) ** 2).sum(axis = 1)))  #subtracting data's columns (overall, potential, etc.) 
    return distances.idxmin(axis=1) 

lable = get_lable(data, centroids)
lable 
```

```python
lable.value_counts()
```

<h2>Update centroids
<h3>find the geometric mean of the features for each player in an assigned cluster

```python
def new_centroids(data, lable, k):
    return data.groupby(lable).apply(lambda x: np.exp(np.log(x).mean())).T

##split the players data frame by cluster (the lable series give cluster assignments).
##to each cluster, calculate the geometric mean of each features (overall, potentials,...)
##the geometric mean is the new centroid 
```

<h2>Repeat step 3+4

```python
from sklearn.decomposition import PCA ##turn 5-dimensional data frame into 2-dimensional df 
import matplotlib.pyplot as plt ##plot the graph
from IPython.display import clear_output ##clear output and give new graph each iteration
```

```python
def plot_cluster(data, lable, centroids, iteration):
    pca = PCA(n_components=2) ##2 columns
    data_2d = pca.fit_transform(data) ##transform
    centroids_2d = pca.fit_transform(centroids.T)
    clear_output(wait=True)
    plt.title(f'Iteration {iteration}')
    plt.scatter(x=data_2d[:,0], y=data_2d[:,1], c=lable)
    plt.scatter(x=centroids_2d[:,0], y=centroids_2d[:,1])
    plt.show()

```

```python
max_iter = 100
k = 5 ##num of cluster

centroids = rand_centroids(data, k)
old_centroids = pd.DataFrame()
iteration = 1 

while iteration < 100 and not centroids.equals(old_centroids): 
    old_centroids = centroids
    lable = get_lable(data, centroids)
    centroids = new_centroids(data, lable, k)
    plot_cluster(data, lable, centroids, iteration)
    iteration += 1 
```
